```python
import cv2
import math
import numpy as np
from rplidar import RPLidar
import matplotlib.pyplot as plt
```

```python
#data scaning
PORT_NAME = 'COM3'
lidar = RPLidar(PORT_NAME)
scan_result = []
for i, scan in enumerate(lidar.iter_scans()):
    scan_result.append(scan)
    if i > 10:
        break
lidar.stop()
lidar.stop_motor()
lidar.disconnect()
```

```python

lidar_data = []
for i in scan_result:
    for j in i:
        if j[0] > 11:
           lidar_data.append([j[2], np.deg2rad(j[1])])
        else:
            continue

# Kích thước canvas
width, height = 1000, 1000
center_x, center_y = width // 2, height // 2

# Khởi tạo hình ảnh
map_img = np.zeros((height, width, 3), dtype=np.uint8)


map_img[center_x - 10 : center_x + 10, center_y - 10 : center_y + 10] = 255


mx, mn = 400, 30

for distance, angle in lidar_data:
    # Chuyển từ cực sang Descartes
    if distance > mx: distance = mx - 1
    if distance < mn: distance = mn - 1
    x = int(distance * np.cos(angle)) + mx
    y = int(distance * np.sin(angle)) + mx 
    # Vẽ điểm (hoặc đường)
    cv2.circle(map_img, (x, y), 3, (0, 255, 0), -1)  # Màu xanh lá

#map_img = cv2.flip(map_img, 0)
cv2.imshow("LiDAR 2D Map", map_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
